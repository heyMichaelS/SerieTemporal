# Clustering - K-Means
> Estes dados são o resultado de uma análise química de vinhos cultivados na mesma região da Itália, derivados de três cultivos diferentes. A análise determinou as quantidades de 13 constituintes encontrados em cada um dos três tipos de vinhos.

### Preprocess

```python
import pandas as pd
# Lemos dados sobre wine
urlWine = 'https://raw.githubusercontent.com/heyMichaelS/SerieTemporal/main/K-Means/wine-clustering.csv'
wine = pd.read_csv(urlWine, sep = ',')
wine.head(5)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(wine)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('Cotovelo')
plt.xlabel('Numero de cluster')
plt.ylabel('CS')
plt.show()
```

```python
# Agrupamos usando K-Means
from sklearn.cluster import KMeans
# Vamos gerar 3 grupos
# Para encontrar a melhor quantidade de grupos veja: https://medium.com/pizzadedados/kmeans-e-metodo-do-cotovelo-94ded9fdf3a9
model = KMeans(n_clusters= 3)
model.fit(wine)
group = model.predict(wine)

wine.head(20)
```

```python
from sklearn.decomposition import PCA

# PCA (Principal Component Analysis: https://builtin.com/data-science/step-step-explanation-principal-component-analysis) com 2 componentes
# O correto seria verificar a melhor quantidade de componentes (exemplo: https://www.kaggle.com/code/jeanc2020/pca-clustering-kmeans)
pca = PCA(n_components=2)
pca_array = pca.fit_transform(wine)

# Converte em dataframe
df_pca = pd.DataFrame(data=pca_array, index=wine.index.copy())
df_pca.columns = ['PC' + str(col+1) for col in df_pca.columns.values]
df_pca['group'] = group

df_pca.head()
```

```python
import matplotlib.pyplot as plt

colors = ['#5DF534', '#425B26', '#151716', '#876E03', '#E8CECF', '#023411', '#042E54', '#2572D3', '#FD0D0C', '#169C90']
df_pca['color'] = df_pca['group'].map({0:colors[0], 1:colors[1], 2:colors[2], 3:colors[3], 4:colors[4], 5:colors[5], 6:colors[6], 7:colors[7], 8:colors[8], 9:colors[9]})

# Plot dos pontos
plt.figure(figsize=(10, 8))
plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df_pca['color'], alpha = 0.6, s=10)

# Calculando e plotando os centróides
centroids = df_pca.groupby('group')[['PC1', 'PC2']].mean()
plt.scatter(centroids['PC1'], centroids['PC2'], c='red', s=100, marker='X', label='Centroids')

# Configurações do gráfico
plt.title('Scatter plot with Centroids')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.grid(True)

# Mostrando o gráfico
plt.show()
```

```python
# Com 3 componentes
pca = PCA(n_components=3)
pca_array = pca.fit_transform(wine)

# Converte em dataframe
df_pca = pd.DataFrame(data=pca_array, index=wine.index.copy())
df_pca.columns = ['PC' + str(col+1) for col in df_pca.columns.values]
df_pca['group'] = group

df_pca.head()
```

```python
# Import libraries
from mpl_toolkits import mplot3d

colors = ['#5DF534', '#425B26', '#151716', '#876E03', '#E8CECF', '#023411', '#042E54', '#2572D3', '#FD0D0C', '#169C90']
df_pca['color'] = df_pca['group'].map({0:colors[0], 1:colors[1], 2:colors[2], 3:colors[3], 4:colors[4], 5:colors[5], 6:colors[6], 7:colors[7], 8:colors[8], 9:colors[9]})

# Criando a figura e o eixo 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.set_title('Agrupamento K-Means')

# Plot dos pontos
scatter = ax.scatter3D(df_pca['PC1'], df_pca['PC2'], df_pca['PC3'], c=df_pca['color'], alpha=0.6, s=10)

# Calculando e plotando os centróides
centroids = df_pca.groupby('group')[['PC1', 'PC2', 'PC3']].mean()
ax.scatter3D(centroids['PC1'], centroids['PC2'], centroids['PC3'], c='red', s=100, marker='X', label='Centroids')

# Configurações do gráfico
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.legend()
ax.grid(True)
```
